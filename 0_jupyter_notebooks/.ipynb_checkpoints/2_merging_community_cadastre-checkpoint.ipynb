# Data Merging (Notebook 2 of 4) 

```python
# Importing libraries
import pandas as pd
```

```python
# Adding the manually labelled data to df, based on the NATIONALCA (Unique Cadatre number

# read 2 files
data1 = pd.read_csv('labelled_data.csv')
data2 = pd.read_csv('output.csv')
```

```python
data1.head(5)
```

```python
data2.head(5)
```

```python
# Select the necessary columns from data1
selected_data1 = data1[['NATIONALCA', 'building_age']]

# Perform the merge baased on NATIONALCA
merged_df = pd.merge(data2, selected_data1, on='NATIONALCA', how='inner')
```

```python
# Save Results
merged_df.to_csv('output_with_types.csv', index=None) 
```
