# Data Merging (Notebook 2 of 4) 

```python
# Import libraries
import pandas as pd
```

```python
# Add the manually labelled data to df, based on the NATIONALCA (Unique Cadastre Reference)

# Read the two files
data1 = pd.read_csv('../2_cleaned_data/a_cleaned_data.csv')
data2 = pd.read_csv('../2_cleaned_data/c_cleaned_data_now_labelled.csv')
```

```python
data1.columns
```

```python
data2.head(5)
```

```python
# Select the necessary columns from data1
selected_data2 = data2[['NATIONALCA', 'building_age']]

# Perform the merge based on NATIONALCA
merged_df = pd.merge(data1, selected_data2, on='NATIONALCA', how='inner')
```

```python
merged_df
```

```python
# Save Results of the merge (merging based on 'NATIONALCA' - national cadastral reference number)
merged_df.to_csv('../2_cleaned_data/d_merged_with_labels.csv', index=None) 
```
